```python
# Required Libraries 
import sympy as sp
```

```python
# Question 4a)

# Defining variables 
x, y = sp.symbols('x y')

# Define a function
f = sp.exp(x)*sp.sin(y)+y**3

# Computing first partial derivative of f with respect to x
df_dx = sp.diff(f, x)

# Computing first partial derivative of f with respect to y
df_dy = sp.diff(f, y)

# Creating output 
print("The derivative of f with respect to x is: df_dx =", df_dx)
print("The derivative of f with respect to y is: df_dy =", df_dy)
```

```python
# Question 4b)

# Defining variables 
x, y = sp.symbols('x y')

# Define a function
g = y*x**2 + x*y**2

# Computing partial derivative of f with respect to x
dg_dx = sp.diff(g, x)

# Computing partial derivative of f with respect to y
dg_dy = sp.diff(g, y)

# Gradient vector at any point
gradident_vector = (dg_dx, dg_dy)

# At the point (1, -1)
gradient_vector = (dg_dx.subs({x: 1, y: -1}), dg_dy.subs({x: 1, y: -1}))

# Finding the magnitude of the gradient vector at (1, -1)
magnitude_gradient_vector = sp.sqrt(gradient_vector[0]**2 + gradient_vector[1]**2)

print("The magnitude of the gradient vector at (1, -1) is", magnitude_gradient_vector)  
```

```python
# Question 4c) 

# Defining variables 
x, y = sp.symbols('x y')

# Define a function
h = sp.ln(x**2 + y**2)

# Computing first partial derivatives 
dh_dx = sp.diff(h, x)
dh_dy = sp.diff(h, y) 

# Computing second partial derivatives 
# note that the '2' is supposed to be exponent 2 
d2h_dx2 = sp.diff(dh_dx, x)
d2h_dy2 = sp.diff(dh_dy, y)

d2h_dxdy = sp.diff(dh_dy, x)
d2h_dydx = sp.diff(dh_dx, y)  

print("d2h_dx2 =", d2h_dx2)
print("d2h_dy2 =", d2h_dy2)
print("d2h_dxdy =", d2h_dxdy) 
print("d2h_dydx =", d2h_dydx) 

# Comments on output
# Since the partial derivatives d2h_dx2 = d2h_dy2 and d2h_dxdy = d2h_dydx, the function is highly symmetrical.  
```

```python
# Question 4d)

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the function
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

# Convert the SymPy expression to a numerical function
j_func = sp.lambdify((x, y), j, 'numpy')

# Create a grid of x and y values and evaluate j on this grid
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

# Plot the contour
plt.contourf(X, Y, Z, levels=50) #The greater the number of levels, the less defined the contour lines become
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
