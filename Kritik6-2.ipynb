```python
# code that integrates the normal_density function between [a,b]

import numpy as np
import matplotlib.pyplot as plt 

mean = 0
variance = 1
a = -10
b = 10

#defining the density function
def normal_density(mean, variance, x):
    f = (1/(np.sqrt(2*np.pi*(variance))))*(np.exp(-((x-mean)**2)/(2*(variance)))) 
    return f 

#defining new integration function
def integration(mean, variance, a,b, n=1000):
    delta_x = (b-a)/n
    area_under_curve = 0 # initializing area to zero, where areas based on xi will be added 

    for i in range(n):
        xi = a + i * delta_x
        yi = normal_density(mean, variance, xi)
        area_under_curve += delta_x * yi

    return area_under_curve

print("The Area under the Normal Density Curve is" , integration(mean, variance, a,b))
```

```python
# code for part a

# arbitrary interval inputs 
a = 1
b = 10

# code for fX(x) function
def uniform_distribution(a, b, x):
    if a <= x <= b:
        return 1/(b-a)
    else:
        return 0

# similar to integration code from Question 3
def expected_value(a, b, n=1000): # no mean and variance
    delta_x = (b-a)/n
    area_under_curve = 0

    for i in range(n):
        xi = a + i * delta_x
        yi = xi * uniform_distribution(a, b, xi) # code for integrant: xfX(x) 
        area_under_curve += delta_x * yi

    return area_under_curve

E_x = expected_value(a, b)

print("The expected value for uniform distribution with a =", a, "and b =", b, "is", E_x)  
```

```python
# code for part b

# arbitrary interval inputs 
a = 1968 # date of last pandemic outbreak of influenza
b = 2020 # date of COVID
rate = 1/50 #rate between pandemics  

# code for fX(x) function
def exponential_distribution(a, b, x):
    if x >= 0:
        return rate*np.exp(-rate*x)
    else:
        return 0

# same as expected_value function from part a, defining the numerical integration from E(x)
def expected_value(a, b, n=1000): # no mean and variance
    delta_x = (b-a)/n
    area_under_curve = 0

    for i in range(n):
        xi = a + i * delta_x
        yi = xi * exponential_distribution(a, b, xi) # code for integrant: xfX(x) 
        area_under_curve += delta_x * yi

    return area_under_curve

E_x = expected_value(a, b)

print("The expected value for exponential distribution with a =", a, ", b =", b, "and a rate of", rate, "is", E_x)
```
