```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return x**2

def gradient_descent(f, learning_rate, initial_point):

  def deriv(f, base_point):
    return(f(base_point+10**(-10))-f(base_point-10**(-10)))/(2*10**(-10))

  x_coords=[initial_point]
  y_coords=[f(initial_point)]

  current_point = initial_point
  max_steps = 1000
  step = 0 #initial step, something for the while loop to start from
  tolerance = 1*10**(-10)

  while step < max_steps:
    gradient = deriv(f, current_point)
    current_point = current_point - learning_rate*gradient
    x_coords.append(current_point)
    y_coords.append(f(current_point))
    step += 1
    if abs(gradient) < tolerance:
      break

  return x_coords, y_coords

learning_rate = 0.000001
initial_point = 0.01
x_coords, y_coords = gradient_descent(f, learning_rate, initial_point)

plot_range=np.linspace(min(x_coords)-0.5, max(x_coords)+0.5, 10000)
function_range=[f(i) for i in plot_range]
plt.plot(plot_range, function_range)
plt.plot(x_coords, y_coords)
plt.show()

print("Minimum found at x =", round(x_coords[-1],3), "and y =", round(y_coords[-1], 3))
```
